# Librosa

```python
import yt_dlp
import os

def download_youtube_audio(video_id, output_path="/code/youssef-benk/vinylitics/raw_data_youtube/songs/"):
    """
    Download the audio from a YouTube video given its ID.
    """

    url = f"https://www.youtube.com/watch?v={video_id}"

    ydl_opts = {
        "format": "bestaudio/best",
        "outtmpl": f"{output_path}{video_id}.mp3",
        "postprocessors": [{"key": "FFmpegExtractAudio", "preferredcodec": "mp3"}]
    }

    with yt_dlp.YoutubeDL(ydl_opts) as ydl:
        ydl.download([url])

    return f"{output_path}{video_id}.mp3"

video_id = "vBynw9Isr28"
audio_file = download_youtube_audio(video_id)
print(f"Downloaded audio file: {audio_file}")
```
