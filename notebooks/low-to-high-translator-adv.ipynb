## Plotting high level features from both datasets

```python
import pandas as pd
import numpy as np

# Suppose df contains the Spotify distribution for acousticness/instrumentalness
spotify_acoustic = df["acousticness"]
spotify_instrumental = df["instrumentalness"]

# --- Find relevant quantiles in Spotify's distribution ---
acous_q_90 = spotify_acoustic.quantile(0.90)
acous_q_95 = spotify_acoustic.quantile(0.95)

instr_q_85 = spotify_instrumental.quantile(0.85)
instr_q_90 = spotify_instrumental.quantile(0.90)
instr_q_95 = spotify_instrumental.quantile(0.95)

# --- Extract the acousticness/instrumentalness columns from echonest ---
acoustic = echonest.xs('acousticness', level=2, axis=1).squeeze()
instrumental = echonest.xs('instrumentalness', level=2, axis=1).squeeze()

# ----------------------------------------------------------------------------------
# STEP 1: Remove 90% of rows where BOTH acousticness & instrumentalness > their 90th pct
# ----------------------------------------------------------------------------------
mask_both_high = (acoustic > acous_q_90) & (instrumental > instr_q_90)
keep_both_high = mask_both_high[mask_both_high].sample(frac=0.1, random_state=42).index
echonest_step1 = echonest[~mask_both_high | echonest.index.isin(keep_both_high)]

# ----------------------------------------------------------------------------------
# STEP 2: Remove 70% of rows where instrumentalness > 85th percentile
#       (trims upper-mid range of instrumentalness)
# ----------------------------------------------------------------------------------
instrumental_step1 = instrumental[echonest_step1.index]
mask_instr_85 = (instrumental_step1 > instr_q_85)
keep_instr_85 = mask_instr_85[mask_instr_85].sample(frac=0.3, random_state=42).index
echonest_step2 = echonest_step1[~mask_instr_85 | echonest_step1.index.isin(keep_instr_85)]

# ----------------------------------------------------------------------------------
# STEP 3: Remove 90% of rows where instrumentalness > 95th percentile
#       (aggressively trim the extreme tail near 1.0)
# ----------------------------------------------------------------------------------
instrumental_step2 = instrumental[echonest_step2.index]
mask_instr_95 = (instrumental_step2 > instr_q_95)
keep_instr_95 = mask_instr_95[mask_instr_95].sample(frac=0.1, random_state=42).index
echonest_step3 = echonest_step2[~mask_instr_95 | echonest_step2.index.isin(keep_instr_95)]

# ----------------------------------------------------------------------------------
# STEP 4: Remove 70% of rows where acousticness > 95th percentile
#       (if acousticness tail is still too big)
# ----------------------------------------------------------------------------------
acoustic_step3 = acoustic[echonest_step3.index]
mask_acous_95 = (acoustic_step3 > acous_q_95)
keep_acous_95 = mask_acous_95[mask_acous_95].sample(frac=0.3, random_state=42).index

# Final filtered DataFrame
echonest_filtered = echonest_step3[~mask_acous_95 | echonest_step3.index.isin(keep_acous_95)]
print("Final shape:", echonest_filtered.shape)
```

## Scaling X and y

```python
y.columns
```

```python
# checking if the correlation between acousticness and instrumentalness is significant

# Filter the DataFrame for cases where either value is over 0.8
filtered_y = y[(y['echonest_audio_features_acousticness'] > 0.8) | (y['echonest_audio_features_instrumentalness'] > 0.8)]

# Calculate the correlation
correlation = filtered_y['echonest_audio_features_acousticness'].corr(filtered_y['echonest_audio_features_instrumentalness'])
print("Correlation between acousticness and instrumentalness (filtered):", correlation)
```

```python
# Adopting and individual approach for each group of columns

from sklearn.preprocessing import QuantileTransformer, MinMaxScaler, PowerTransformer, RobustScaler


# Columns that tend to be heavily skewed
skewed_cols = ['echonest_audio_features_liveness', 'echonest_audio_features_speechiness']

# Columns with extreme values on either end
u_shaped_cols = ['echonest_audio_features_instrumentalness', 'echonest_audio_features_acousticness']

# More evenly distributed and bounded columns
even_cols =['echonest_audio_features_energy', 'echonest_audio_features_danceability', 'echonest_audio_features_valence']

# Create separate transformer instances
skewed_transformer = PowerTransformer(method='yeo-johnson')
u_shaped_transformer = QuantileTransformer(output_distribution='uniform')
even_transformer = MinMaxScaler()

# --- Fit transformers on their respective columns ---
skewed_transformer.fit(y[skewed_cols])
u_shaped_transformer.fit(y[u_shaped_cols])
even_transformer.fit(y[even_cols])

# --- Transform each group separately ---
y_skewed_scaled    = skewed_transformer.transform(y[skewed_cols])
y_u_shaped_scaled  = u_shaped_transformer.transform(y[u_shaped_cols])
y_even_scaled      = even_transformer.transform(y[even_cols])

# --- Desired final order ---
final_order = [
    'echonest_audio_features_acousticness',    # from u_shaped
    'echonest_audio_features_danceability',      # from even
    'echonest_audio_features_energy',            # from even
    'echonest_audio_features_instrumentalness',    # from u_shaped
    'echonest_audio_features_liveness',          # from skewed
    'echonest_audio_features_speechiness',       # from skewed
    'echonest_audio_features_valence'            # from even
]

y_skewed_scaled_df = pd.DataFrame(y_skewed_scaled, columns=skewed_cols, index=y.index)
y_u_shaped_scaled_df = pd.DataFrame(y_u_shaped_scaled, columns=u_shaped_cols, index=y.index)
y_even_scaled_df = pd.DataFrame(y_even_scaled, columns=even_cols, index=y.index)

# Merge all scaled DataFrames
y_scaled_merged = pd.concat([y_u_shaped_scaled_df, y_even_scaled_df, y_skewed_scaled_df], axis=1)
# Reorder the columns to match the final order
y_scaled = y_scaled_merged[final_order]

# --- Fit a scaler on the entire X DataFrame ---
x_scaler = RobustScaler()
x_scaler.fit(X)
X_scaled = pd.DataFrame(x_scaler.transform(X), columns=X.columns, index=X.index)
```

## Reduce dimensionality

```python
import numpy as np
from sklearn.decomposition import PCA

pca = PCA(n_components=100)
X_reduced = pca.fit_transform(X_scaled)
print("Total variance explained:", np.sum(pca.explained_variance_ratio_))
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_reduced, y_scaled, test_size=0.3, random_state=42)
print("Training set shape:", X_train.shape, y_train.shape)
print("Testing set shape:", X_test.shape, y_test.shape)
```

# Dense Neural Network

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.regularizers import l2

input_dim = X_train.shape[1]
output_dim = y_train.shape[1]
regularization = l2(0.01)

model = Sequential([
    Dense(256, activation='relu', input_shape=(input_dim,), kernel_regularizer=regularization),
    Dropout(0.2),
    Dense(128, activation='relu', kernel_regularizer=regularization),
    Dense(64, activation='relu', kernel_regularizer=regularization),
    Dense(64, activation='relu', kernel_regularizer=regularization),
    Dense(output_dim, activation='linear', kernel_regularizer=regularization)
])

model.compile(optimizer=Adam(learning_rate=1e-3), loss='mse', metrics=['mae'])
model.summary()
```

```python
history = model.fit(X_train, y_train, epochs=50, batch_size=32, validation_split=0.2)
```

```python
loss, mae = model.evaluate(X_test, y_test)
print("Test Loss (MSE):", loss)
print("Test MAE:", mae)
```
