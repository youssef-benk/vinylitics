**Cleaning the data**

```python
#delete unnamed:0 column --> redundant with index
df_cleaned = data.drop(columns= 'Unnamed: 0')
df_cleaned
```

```python
#remove track name, album names, etc.. maybe can lead to overfitting if we keep everything
df_cleaned = df_cleaned.drop(columns=["artists", "album_name", "track_name"]) #voir si on doit remettre ou enlever des colonnes
df_cleaned
```

```python
#Convert 'explicit' from boolean to numerical (0 = False, 1 = True)

df_cleaned["explicit"] = df_cleaned["explicit"].astype(int)
df_cleaned.head(30)
```

```python
#Song length from ms to minute or seconds maybe --> voir si peut être drop
df_cleaned["song_length_minutes"] = df_cleaned["duration_ms"] / 60000
df_cleaned
```
