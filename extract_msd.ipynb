```python
import os
import h5py
import pandas as pd


# Function to safely decode bytes to string
def safe_decode(value):
    if isinstance(value, bytes):
        return value.decode("utf-8")
    return value  # Return as-is if not bytes

# Function to safely convert to float
def safe_float(value):
    try:
        return float(value)
    except (ValueError, TypeError):
        return None

# Function to safely convert to int
def safe_int(value):
    try:
        return int(value)
    except (ValueError, TypeError):
        return None

# Function to extract data from a single .h5 file
def extract_song_data(h5_file):
    try:
        with h5py.File(h5_file, 'r') as f:
            metadata = f['metadata']['songs'][0]
            analysis = f['analysis']['songs'][0]
            musicbrainz = f['musicbrainz']['songs'][0] if len(f['musicbrainz']['songs']) > 0 else None

            song_data = {
                # ‚úÖ Fixed Metadata Extraction
                "track_id": safe_decode(analysis[30]),  # Track ID
                "title": safe_decode(metadata[18]),  # Title
                "artist_name": safe_decode(metadata[9]),  # Artist Name
                "album_name": safe_decode(metadata[14]),  # Album Name
                "artist_id": safe_decode(metadata[8]),  # Artist ID

                # ‚úÖ Fixed MusicBrainz Data
                "release_year": safe_int(musicbrainz[0]) if musicbrainz else None,

                # ‚úÖ Fixed Artist Scores
                "artist_familiarity": safe_float(metadata[2]),
                "artist_hotttnesss": safe_float(metadata[3]),

                # ‚úÖ Fixed Audio Features
                "tempo": safe_float(analysis[27]),  # Tempo (BPM)
                "key": safe_int(analysis[28]),  # Musical Key
                "mode": safe_int(analysis[24]),  # Major/Minor Mode
                "time_signature": safe_int(analysis[22]),  # Time Signature (Integer)
                "loudness": safe_float(analysis[23]),  # Loudness (dB)
                "duration": safe_float(analysis[26]),  # Duration (Seconds)
                "end_of_fade_in": safe_float(analysis[4]),
                "start_of_fade_out": safe_float(analysis[5]),
            }

            return song_data

    except Exception as e:
        print(f"‚ùå Error processing file {h5_file}: {e}")
        return None

# Loop through all .h5 files and extract data
song_list = []
for root, dirs, files in os.walk(DATASET_PATH):
    for file in files:
        if file.endswith(".h5"):
            file_path = os.path.join(root, file)
            song_data = extract_song_data(file_path)
            if song_data:
                song_list.append(song_data)

# Convert to DataFrame and save as CSV
df = pd.DataFrame(song_list)
df.to_csv("million_song_all_features_final.csv", index=False)
print("‚úÖ Full dataset extracted and saved as CSV!")
```
